```python
columns=['발명의명칭']
text=kipris[columns]
```

```python
text
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
 !pip install soynlp
```

```python
from soynlp.tokenizer import RegexTokenizer
```

```python
tokenizer = RegexTokenizer()
tokenizer
```

```python
kipris_text = text["발명의명칭"].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```python
kipris_text
```

```python
# 실해팼지만 시도하면서 봤던 블로그들
# https://monetd.github.io/python/nlp/Word-Embedding-Word2Vec-%EC%8B%A4%EC%8A%B5/
# https://wikidocs.net/21698
# https://wonhwa.tistory.com/23
# https://soy3on.tistory.com/177
# https://hanshuginn.blogspot.com/2019/03/python-doc2vec.html
# https://colab.research.google.com/github/yooseonghwan/OpenDataWrangling/blob/master/03_word2vec%EC%9C%BC%EB%A1%9C_%EB%8B%A8%EC%96%B4%EC%9C%A0%EC%82%AC%EB%8F%84_%EB%B3%B4%EA%B8%B0_teacher.ipynb#scrollTo=p6LvRooFTnrp
# 유튜브 https://www.youtube.com/watch?v=5ivVf-Guqk4&list=PLrLEKGJAgXxL-R9IqDH7HANWXRsS900tF&index=2 
```

```python
# https://perconsi.tistory.com/87 보고 했습니다.
# 텍스트 데이터 토큰화 
%time tokens = kipris_text.apply(tokenizer.tokenize)
tokens[:]
```

```python
# 로그를 출력하기 위해 불러오기
import logging
logging.basicConfig(
    format='%(asctime)s : %(levelname)s : %(message)s', 
    level=logging.INFO)
```

```python
from gensim.models import word2vec
model = word2vec.Word2Vec(tokens, min_count=2, size=300, window=5, iter=10)
model
```

```python
# 모델 이름을 지정하고 저장한다.
model_name = 'kiprisword2vec'
model.save(model_name)
```

```python
# 단어 사전 수
len(model.wv.vocab)
```

```python
# 단어 사전 top 뽑기
vocab = model.wv.vocab
sorted(vocab, key=vocab.get, reverse=True)[:30]
```

```python
# Counter로 자주 등장하는 단어 보기
from collections import Counter
dict(Counter(vocab).most_common(50))
```
